#importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas DataFrame
df=pd.read_csv("/content/credit_data.csv")
```

This datast highly unbalanced

```python
#0 => Normal Transaction
#1 => Fraudulent Transaction
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
fraud.Amount.describe()
#
```

```python
# compare the values for both transactions
df.groupby('Class').mean()
```

Undersampling

Build a sample datatset containing the normal transaction and Fraud transaction

Fraud Transaction -492


```python
legit_sample=legit.sample(n=492)
```

Concatenating the dataframe

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

#Splitting the data and label


```python
X=new_df.drop(columns='Class',axis=1)
Y=new_df['Class']
```

```python
print(X)
print(Y)
```

Splitting the training and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

Model Training

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

#Model Evaluation

Accuracy Score

```python
#accuracy Score on train
X_train_prediction=model.predict(X_train)
X_train_accuracy=accuracy_score(Y_train,X_train_prediction)
```

```python
print(X_train_accuracy)
```

```python
#accuracy Score on test
X_test_prediction=model.predict(X_test)
X_test_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python
print(X_test_accuracy)
```
